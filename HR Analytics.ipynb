Checking for Outliers

```python
#Check the outliers in continuous data columns
plt.figure(figsize=(12,18),facecolor="white")
plotnumber=1
for col in continuous_col:
    ax=plt.subplot(5,3,plotnumber)
    sns.boxplot(df[col],palette="bright")
    plt.xlabel(col,fontsize=15)
    plt.yticks(rotation=0,fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

Observations:

We can find outliers in the features "MonthlyIncome", "NumCompaniesWorked", "TotalWorkingYears", "TrainingTimesLastYear", "YearsAtCompany", "YearsInCurrentRole", "YearsSinceLastPromotion" and "YearsWithCurrManager". We will remove these outliers by using Z-Score technique.

```python
#Remove outliers
from scipy.stats import zscore
out_features=df[["MonthlyIncome","NumCompaniesWorked","TotalWorkingYears","TrainingTimesLastYear","YearsAtCompany","YearsInCurrentRole","YearsSinceLastPromotion","YearsWithCurrManager"]]
z=np.abs(zscore(out_features))
df1=df[(z<3).all(axis=1)]
```

```python
df1.shape
```

```python
print("Data loss percentage - ",((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

The data loss percentage is 5.64% which is less than 10% and hence acceptable.

Encoding Categorical Columns

```python
#Separate columns with object data type
obj_col=[]
for col in df1.columns:
    if df1[col].dtype==object:
        obj_col.append(col)
obj_col
```

```python
from sklearn.preprocessing import OrdinalEncoder
OE=OrdinalEncoder()
for col in obj_col:
    df1[col]=OE.fit_transform(df1[col].values.reshape(-1,1))
df1
```

Checking for Skewness

```python
df1.skew()
```

The continuous features "DistanceFromHome", "MonthlyIncome", "NumCompaniesWorked", "PercentSalaryHike", "TotalWorkingYears", "TrainingTimesLastYear", "YearsAtCompany", "YearsInCurrentRole", "YearsSinceLastPromotion" and "YearsWithCurrManager" have skewness. We will remove the skewness using Yeo-Johnson transformation method.

```python
#Separate columns with skewness to be removed.
skewed_col=["DistanceFromHome","MonthlyIncome","NumCompaniesWorked","PercentSalaryHike","TotalWorkingYears","TrainingTimesLastYear","YearsAtCompany","YearsInCurrentRole","YearsSinceLastPromotion","YearsWithCurrManager"]
skewed_col
```

```python
#Remove skewness
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson')
for col in skewed_col:
    df1[col]=pt.fit_transform(df1[[col]])
```

Separating features and label

```python
x=df1.drop('Attrition',axis=1)
y=df1['Attrition']
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

Now, we have done feature scaling to overcome the issue of biasness in the model.

Checking Variance Inflation Factor (VIF)

```python
#Find out Variance Inflation Factor of each feature
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif["VIF Values"]=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif["Features"]=x.columns
vif
```

We check VIF values to find out if there is any multicollinearity problem with the features which we need to solve. In the above table, we can see that none of the features have VIF values greater than 10. So, we can conclude that there is no multicollinearity problem with the features and hence we can proceed further.

```python
y.value_counts()
```

We can see from the above values that the data in the label is not balanced. As this is a classification problem, we need to balance the data and we will do this by oversampling the data using SMOTE method.

Balancing the Label

```python
#Oversampling the data using SMOTE
from imblearn.over_sampling import SMOTE
SM=SMOTE()
x1,y1=SM.fit_resample(x,y)
```

# Model Building

Finding the best random state

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.30,random_state=i)
    RFC=RandomForestClassifier()
    RFC.fit(x_train,y_train)
    pred=RFC.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is ",maxAccu,"at random state",maxRS)
```

Creating train test split

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.30,random_state=maxRS)
```

Importing the necessary modules

```python
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import ExtraTreesClassifier,GradientBoostingClassifier,BaggingClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#Random Forest Classfier
RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)
predRFC=RFC.predict(x_test)
print(accuracy_score(y_test,predRFC))
print(confusion_matrix(y_test,predRFC))
print(classification_report(y_test,predRFC))
```

```python
#Logistic Regression
LR=LogisticRegression()
LR.fit(x_train,y_train)
predLR=LR.predict(x_test)
print(accuracy_score(y_test,predLR))
print(confusion_matrix(y_test,predLR))
print(classification_report(y_test,predLR))
```

```python
#XGB Classifier
XGB=XGBClassifier()
XGB.fit(x_train,y_train)
predXGB=XGB.predict(x_test)
print(accuracy_score(y_test,predXGB))
print(confusion_matrix(y_test,predXGB))
print(classification_report(y_test,predXGB))
```

```python
#Decision Tree Classifier
DTC=DecisionTreeClassifier()
DTC.fit(x_train,y_train)
predDTC=DTC.predict(x_test)
print(accuracy_score(y_test,predDTC))
print(confusion_matrix(y_test,predDTC))
print(classification_report(y_test,predDTC))
```

```python
#Support Vector Machine Classifier
svc=SVC()
svc.fit(x_train,y_train)
predsvc=svc.predict(x_test)
print(accuracy_score(y_test,predsvc))
print(confusion_matrix(y_test,predsvc))
print(classification_report(y_test,predsvc))
```

```python
#Extra Trees Classifier
ET=ExtraTreesClassifier()
ET.fit(x_train,y_train)
predET=ET.predict(x_test)
print(accuracy_score(y_test,predET))
print(confusion_matrix(y_test,predET))
print(classification_report(y_test,predET))
```

```python
#Gradient Boosting Classifier
GB=GradientBoostingClassifier()
GB.fit(x_train,y_train)
predGB=GB.predict(x_test)
print(accuracy_score(y_test,predGB))
print(confusion_matrix(y_test,predGB))
print(classification_report(y_test,predGB))
```

```python
#Bagging Classifier
BC=BaggingClassifier()
BC.fit(x_train,y_train)
predBC=BC.predict(x_test)
print(accuracy_score(y_test,predBC))
print(confusion_matrix(y_test,predBC))
print(classification_report(y_test,predBC))
```

# Hyper Parameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from scipy.stats import expon,randint
param={'C': expon(scale=100), # Exponential distribution for C parameter
       'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],
       'gamma': expon(scale=0.1), # Exponential distribution for gamma parameter
       'degree': randint(2, 5), # Integer values for degree parameter (for poly kernel)
       'coef0': [0.0, 0.1, 0.5, 1.0] # Possible values for coef0 (for poly and sigmoid kernels)
      }
RCV=RandomizedSearchCV(SVC(),param,cv=5)
RCV.fit(x_train,y_train)
```

```python
#Best parameters for the final model
RCV.best_params_
```

# Building the Final Model

```python
final_model=SVC(C=35.245626230755924,gamma=0.09843056892379844,kernel='rbf',coef0=0.0,degree=2)
final_model.fit(x_train,y_train)
pred=final_model.predict(x_test)
acc=accuracy_score(y_test,pred)
print("Accuracy acheived by the final model is ",acc*100)
```

Plotting ROC and comparing AUC for the final model

```python
from sklearn import metrics
fpr,tpr,thresholds=metrics.roc_curve(y_test,pred)
roc_auc=metrics.auc(fpr,tpr)
display=metrics.RocCurveDisplay(fpr=fpr,tpr=tpr,roc_auc=roc_auc,estimator_name=final_model)
display.plot()
plt.show()
```

Above is the ROC curve for the final model and AUC value for the final model is 98%.
